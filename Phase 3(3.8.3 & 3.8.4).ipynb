# 3.8.3	Store the data (the result of 3.8.2) in DataFrame and analyze districts that appear 3 or more than 3 years and print the corresponding state/ut, district, murders, and year in descending order.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### import stored data of 3.8.2

```python
df=pd.read_csv("phase-2(3.8.2).csv")
df
```

We have stored the data in the dataframe 'df'

```python
filtered_df = df.groupby('DISTRICT').filter(lambda x: len(x) >= 3)
```

```python
result_df = filtered_df.groupby(['STATE/UT', 'DISTRICT', 'Year'])['Highest_no_Murders'].sum().reset_index()
result_df = result_df.sort_values(by='Highest_no_Murders', ascending=False)
result_df
```

#### this dataframe indicates that the "district"  appears 3 or more than 3 years in the every district and state by Decending order

# 3.8.4	Use appropriate graphs to show your data (the result of 3.8.3)

```python
df_group=result_df.groupby(by='STATE/UT').sum().reset_index().sort_values(by='Highest_no_Murders',ascending=False)
df_group
```

```python
plt.figure(figsize=(15,4))
sns.barplot(data=df_group,x='STATE/UT',y='Highest_no_Murders',errwidth=0,palette='coolwarm')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6)) 
plt.scatter(df_group['STATE/UT'], df_group['Highest_no_Murders'], color='b', marker='o', label='Highest_no_Murders')

plt.xlabel('STATE/UT')
plt.ylabel('Highest_no_Murders')
plt.title('Scatter Plot: STATE/UT vs. Highest_no_Murders')
plt.grid(True)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
x=result_df.groupby(by='Year').sum().reset_index().sort_values(by='Highest_no_Murders',ascending=False)

plt.figure(figsize=(15,8))
sns.lineplot(x=x['Year'],y=x['Highest_no_Murders'])
plt.xlabel("DIFFERENT YEARS RANGING FROM 2001 TO 2012")
plt.xticks([x for x in range(2001,2013)])
plt.ylabel("TOTAL NUMBER OF MURDERS!!")
plt.grid()
plt.title("TOTAL MURDER TREND THROUGH DIFFERENT YEARS!!")
plt.show()
```

In the year 2007 recods the lowest no of murders where as highest is in the year 2001

```python
plt.figure(figsize=(18,8))
sns.scatterplot(x='Year',y='Highest_no_Murders',data=result_df,hue='STATE/UT',palette='tab10',style='STATE/UT')
plt.legend(loc='upper right', bbox_to_anchor=(0.7, -0.1),
          fancybox=True, shadow=True, ncol=5)
```

From above we see that in every year Bihar tops the cahrt with most murders and secures the first place.
